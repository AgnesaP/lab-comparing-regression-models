### 3.Great! Now define a function that takes a list of models and train (and tests) them so we can try a lot of them without repeating code.

```python

# defining a function that accepts a list of models and the data & trains and tests the models

def regressors(function_names, X_train,y_train, X_test,y_test):
   
    
    for reg in function_names:
     
  
        print("\n")
      

        reg.fit(X_train,y_train)

        y_train_pred = reg.predict(X_train)
        y_test_pred = reg.predict(X_test)
        predictions_train = reg.predict(X_train)
        predictions_test = reg.predict(X_test)
        
        print("Function: " + str(reg))
        print("R2:")
        display(r2_score(y_train,predictions_train))
        display(r2_score(y_test,predictions_test))
        
        print("RMSE:")
        display(math.sqrt(mean_squared_error(y_train,predictions_train)))
        display(math.sqrt(mean_squared_error(y_test,predictions_test)))
        
        print("MSE:")
        display(mean_squared_error(y_train,predictions_train))
        display(mean_squared_error(y_test,predictions_test))
     
```

```python

#input the models to be used and their parameters

function_names = [
    LinearRegression(),
    KNeighborsRegressor(n_neighbors=5),
    MLPRegressor(hidden_layer_sizes=(150,100,50),  max_iter = 300,activation = 'relu', solver = 'adam')
    ]

#calling function to run the models from function_names as well as the split data
regressors(function_names, X_train,y_train, X_test,y_test)
```

##### Based on the results above, we see that MLPRegressor is giving us the highest R2 score and the lowest RMSE and MSE, which makes it the best option of the three.
